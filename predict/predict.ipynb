```python
def load_preprocessed_image(image_path):
    target_size = (50, 50)
    image = load_img(image_path, target_size=target_size)           # Ridimensiona l'immagine
    image_array = img_to_array(image)                               # Converte l'immagine in un array numpy
    image_array = np.expand_dims(image_array, axis=0)               # Aggiunge una dimensione per il batch
    image_array = image_array / 255.0 
    return image_array
```
