```python
import cv2
import clahe as clahe
import fuzzy as fuzzy
import wavelet as wavelet
import fourier as fourier
```

```python
PATH = '../../images\cancer\8975_idx5_x1501_y1401_class1.png'
```

```python
def plot_images(no_denoised_img, fuzzy_img, wavelet_img, fourier_img):
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))

    axes[0].imshow(no_denoised_img)
    axes[0].set_title('Original')
    axes[1].imshow(fuzzy_img)
    axes[1].set_title('Fuzzy')
    axes[2].imshow(wavelet_img)
    axes[2].set_title('Wavelet')
    axes[3].imshow(fourier_img)
    axes[3].set_title('Fourier')

    for ax in axes:
        ax.axis('off')
    
    plt.tight_layout()
    fig.savefig("compare.jpg")
    plt.show()    
```

```python
no_denoised_img = cv2.imread(PATH)
fuzzy_img = fuzzy.FuzzyContrastEnhance(PATH, target_size=(50, 50, 3), show_result=0, show_histogram=0)
wavelet_img = wavelet.apply_wavelet_denoising(PATH, target_size=(50, 50, 3), debug=0, show_result=0, show_histogram=0)
fourier_img = fourier.apply_fourier_denoising(PATH, target_size=(50, 50, 3), histogram=0, show_result=0)
```

```python
plot_images(no_denoised_img, fuzzy_img, wavelet_img, fourier_img)
```
