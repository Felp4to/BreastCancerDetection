```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Normalization

```python
def apply_normalization(image):
    #image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Normalize image
    normalized_image = cv2.normalize(image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

    # Imposta la figura per mostrare le immagini affiancate
    plt.figure(figsize=(10, 5))

    # Mostra l'immagine originale
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title('Immagine Originale')
    plt.axis('off')

    # Mostra l'immagine normalizzata
    plt.subplot(1, 2, 2)
    plt.imshow(normalized_image.astype(np.uint8))  # Riconverti i pixel a interi
    plt.title('Immagine Normalizzata')
    plt.axis('off')
    
    return normalized_image


def apply_normalization2(image):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # ????

    # Normalize image
    normalized_image = cv2.normalize(image_rgb, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
    
    return normalized_image

```

```python
PATH_IMAGE = './1-126.jpg'
image = cv2.imread(PATH_IMAGE)
apply_normalization(image)
```
