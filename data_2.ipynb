```python
import os
import pandas as pd
from PIL import Image
```

```python
PATH_CSV = "../Datasets/CBIS-DDSM/archive/csv"
PATH_IMAGES = "../Datasets/CBIS-DDSM/archive/jpeg"
CALC_CASE_DESC_TEST_SET_CSV = os.path.join(PATH_CSV, "calc_case_description_test_set.csv")
CALC_CASE_DESC_TRAIN_SET_CSV = os.path.join(PATH_CSV, "calc_case_description_train_set.csv")
DICOM_INFO_CSV = os.path.join(PATH_CSV, "dicom_info.csv")
MASS_CASE_DESC_TEST_SET_CSV = os.path.join(PATH_CSV, "mass_case_description_test_set.csv")
MASS_CASE_DESC_TRAIN_SET_CSV = os.path.join(PATH_CSV, "mass_case_description_train_set.csv")
META_CSV = os.path.join(PATH_CSV, "meta.csv")
```

```python
CLASSES = {0 : 'MALIGNANT', 1:  'BENIGN', 2: 'BENIGN_WITHOUT_CALLBACK'}
```

```python
df_train = pd.read_csv(CALC_CASE_DESC_TRAIN_SET_CSV)
#display(df_train)

df_test = pd.read_csv(CALC_CASE_DESC_TEST_SET_CSV)
display(df_test)
```

```python
df_train.drop('cropped image file path', axis = 1, inplace=True)
df_train.drop('ROI mask file path', axis = 1, inplace=True)

df_test.drop('cropped image file path', axis=1, inplace=True)
df_test.drop('ROI mask file path', axis=1, inplace=True)

def tronca(stringa):
    return stringa.split('/')[-2]

df_train['image file path'] = df_train['image file path'].apply(tronca)
df_test['image file path'] = df_test['image file path'].apply(tronca)

display(df_train)
display(df_test)
```

```python
# mostra mammography

path = os.path.join(PATH_IMAGES, df_train.loc[2, 'image file path'])
path
```

```python
# Cerca la prima immagine JPEG nella cartella
for filename in os.listdir(path):
    if filename.lower().endswith(('.jpeg', '.jpg')):
        # Costruisci il percorso completo del file
        img_path = os.path.join(path, filename)
        # Apri e mostra l'immagine
        img = Image.open(img_path)
        img.show()
        break  # Esci dal ciclo dopo aver mostrato la prima immagine
```
